```python
#Leer un archivo en tu drive
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Ing_caracteristicas/reto_ing_caracteristicas.csv')
df.head()
```

```python
EmpleadosAttrition = df
EmpleadosAttrition.columns.tolist()
```

```python
EmpleadosAttrition.dtypes
```

```python
#Eliminar las columnas que no tiene relación con los cálculos
EmpleadosAttrition.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'], axis = 1, inplace = True)
```

```python
#Calcular los años desde la contratación

#Transformar la columna de fecha en tipo date

EmpleadosAttrition['Year'] = EmpleadosAttrition['HiringDate'].str[-4:]
```

```python
#Crear columna de años en la compañia hasta el 2018

EmpleadosAttrition['Year'] = EmpleadosAttrition['Year'].astype(int)
EmpleadosAttrition['YearsAtCompany'] = 2018 - EmpleadosAttrition['Year']
```

```python
#Renombrar la variable DistanceFromHome

EmpleadosAttrition['DistanceFromHome_km'] = EmpleadosAttrition['DistanceFromHome']

#Crear la nueva variable distanceFromHome que solo contenga numeros

EmpleadosAttrition['DistanceFromHome']=EmpleadosAttrition['DistanceFromHome_km'].str[:-2]
EmpleadosAttrition['DistanceFromHome']=EmpleadosAttrition['DistanceFromHome'].astype(int)
```

```python
#Generar una tabla con los sueldos promedio
SueldoPromedioDepto = EmpleadosAttrition.groupby('Department')['MonthlyIncome'].mean()
SueldoPromedioDepto
```

```python
#Normalizar la variable de Monthly income

#Importar las librerias
from sklearn.preprocessing import MinMaxScaler

#Iniciar el MinMaxScaler
scaler = MinMaxScaler()

#Escalar la variable
EmpleadosAttrition['MonthlyIncome_norm']=scaler.fit_transform(EmpleadosAttrition[['MonthlyIncome']])
```

```python
#Comprobar que funciono la normalizacion
EmpleadosAttrition['MonthlyIncome_norm'].describe()
```

```python
#Convertir la variable categorica en numérica

#Importar OneHotEncode

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)

#Encode Business Travel
encoded_businesstravel = encoder.fit_transform(EmpleadosAttrition[['BusinessTravel']])
encoded_df_businesstravel = pd.DataFrame(encoded_businesstravel, columns=encoder.get_feature_names_out(['BusinessTravel']))

#Encode Department
encoded_department = encoder.fit_transform(EmpleadosAttrition[['Department']])
encoded_df_department = pd.DataFrame(encoded_department, columns=encoder.get_feature_names_out(['Department']))

#Encode EducationField
encoded_educationfield = encoder.fit_transform(EmpleadosAttrition[['EducationField']])
encoded_df_educationfield = pd.DataFrame(encoded_educationfield, columns=encoder.get_feature_names_out(['EducationField']))

#Encode Gender
encoded_gender = encoder.fit_transform(EmpleadosAttrition[['Gender']])
encoded_df_gender = pd.DataFrame(encoded_gender, columns=encoder.get_feature_names_out(['Gender']))

#Encode JobRole
encoded_jobrole = encoder.fit_transform(EmpleadosAttrition[['JobRole']])
encoded_df_jobrole = pd.DataFrame(encoded_jobrole, columns=encoder.get_feature_names_out(['JobRole']))

#Encode MaritalStatus
encoded_maritalstatus = encoder.fit_transform(EmpleadosAttrition[['MaritalStatus']])
encoded_df_maritalstatus = pd.DataFrame(encoded_maritalstatus, columns=encoder.get_feature_names_out(['MaritalStatus']))

#Encode Attrition
encoded_attrition = encoder.fit_transform(EmpleadosAttrition[['Attrition']])
encoded_df_attrition = pd.DataFrame(encoded_attrition, columns=encoder.get_feature_names_out(['Attrition']))
```

```python
#Unirlos al Dataframe

EmpleadosAttrition = pd.concat([EmpleadosAttrition,
                                encoded_df_businesstravel,
                                encoded_df_department,
                                encoded_df_educationfield,
                                encoded_df_gender,
                                encoded_df_jobrole,
                                encoded_df_maritalstatus,
                                encoded_df_attrition], axis = 1)

```

```python
EmpleadosAttrition_df=EmpleadosAttrition.drop(['BusinessTravel','Department','EducationField',
                                               'Gender','JobRole','MaritalStatus','Attrition',
                                               'HiringDate','OverTime','DistanceFromHome_km',
                                               'Attrition_No'],axis=1)
```

```python
EmpleadosAttrition_df
```

```python
corr_matrix = EmpleadosAttrition_df.corr()
```

```python
corr_matrix['Attrition_Yes'].sort_values(ascending = False)
```

```python
selected_variables = corr_matrix [corr_matrix['Attrition_Yes']>=0.1]
selected_variables = selected_variables['Attrition_Yes']
```

```python
selected_variables= selected_variables.index
```

```python
EmpleadosAttritionFinal = EmpleadosAttrition[selected_variables]
EmpleadosAttritionFinal
```
