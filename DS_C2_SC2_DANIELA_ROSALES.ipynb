```python
import numpy as np
import pandas as pd
```

```python

#Extraer la información del archivo
df = pd.read_csv('/content/country_vaccinations.csv')
df.head()
```

```python
#Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, 
#asegurandote que las columnas con fechas sean del tipo datetime64

#Transformar tipo de data para la columna "date"
df['date']=pd.to_datetime(df['date'])

df.dtypes
```

```python
#Determinar la cantidad de vacunas aplicadas de cada compañía (en base a como lo reporta cada país en la columna vaccines)
df.groupby('vaccines').agg({'daily_vaccinations':'sum'})
```

```python
#Obtener la cantidad de vacunas aplicadas en todo el mundo
df['daily_vaccinations'].sum()
```

```python
#Calcular el promedio de vacunas aplicadas por país
(df['daily_vaccinations'].sum())/(df['country'].count())
```

```python
#Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo
df_april = df[df['date']=='2021-01-29']
df_april['daily_vaccinations'].sum()
```

```python
#Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) 
#con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.  

conDiferencias = df
conDiferencias['vaccination_diff'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']

conDiferencias['vaccination_diff'].sum()
```

```python
#Obtener el período de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua

time_diff = df['date'].max() - df['date'].min()
days = time_diff.days

print (f'El dataset contiene infomación para {days} días')
```

```python
#Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y
#una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la
#columna vaccines y separar por el carácter / )

conCantidad = df

conCantidad['canVac'] = conCantidad['vaccines'].str.split(',').str.len()

conCantidad.head()
```

```python
#Generar un dataframe denominado antes20 con todos los registros que se hayan
#realizado antes del 20 de diciembre de 2020.

antes20 = df[df['date']<='2020-12-20']
antes20
```

```python
#Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado
#la vacuna Pfizer

pfizer = df[df['vaccines'].str.contains('Pfizer',case = False)]

pfizer
```

```python
#Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, 
#donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter. 

writer = pd.ExcelWriter("resultadosReto.xlsx")

conDiferencias.to_excel(writer, sheet_name="Con_Diferencias")
conCantidad.to_excel(writer, sheet_name="Con_Cantidad")
antes20.to_excel(writer, sheet_name="antes20")
pfizer.to_excel(writer, sheet_name = "Pfizer")
```
